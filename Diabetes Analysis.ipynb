# Handling Redundancy Example

*******Please note that since my dataset does not have high correlation, following code will not drop any features!*******

```python
high_correlated = [(i,k) for i in range(matrix.shape[1]) for k in range(i + 1, matrix.shape[1]) if abs(matrix.iloc[i,k]) > 0.6]

features_drop = set()
for p in high_correlated:
    features_drop.add(p[1])
    
not_redundant = data.drop(columns=features_drop, axis=1)
not_redundant.info()
```

# Setting the prediction target

```python
x = data[data.columns[1:]]
```

```python
y = data['Diabetes_binary']
```

# Splitting our data as Train Test, Test size being 20%

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(x, y, test_size = 0.2, random_state = 42)
```

# Let's check the shapes or our Train&Test data

```python
print("Train data of X: ", X_train.shape)
print("Train data of Y: ", Y_train.shape)
```

```python
print("Test data of X: ", X_test.shape)
print("Test data of Y: ", Y_test.shape)
```

# Scaling our test and train datasets for performance purposes

```python
scaler = StandardScaler()
X_train_Scaled = scaler.fit_transform(X_train)
X_test_Scaled = scaler.transform(X_test)
```

# Using Logistic Regression Machine Learning Algorithm on our dataset and performing Recursive Feature Elimination for High Scale Dimensionality

*******Even though my dataset doesn't have high scale dimensionality, I wanted to show how it can be done and it also leads to better performance and model's efficiency*******

```python
from sklearn.linear_model import LogisticRegression
from sklearn.feature_selection import RFE

regressor = LogisticRegression()
r_f_e = RFE(regressor, n_features_to_select = 6)
X_train_RecursiveFE = r_f_e.fit_transform(X_train_Scaled, Y_train)
X_test_RecursiveFE = r_f_e.transform(X_test_Scaled)

regressor.fit(X_train_RecursiveFE, Y_train)
```

# As we can see from here, even after dropping the columns in data cleaning part, the highest priorities belonged to the following features. I used my knowledge to drop features in the beginning and analyzed the dataset by myself. As we can see from here, if you do not drop those features, they still wouldn't affect my model. This is a way to handle irrelevant features also. By looking at this, scientists may end up in different results depending on their domain knowledge.

```python
selected_features = X_train.columns[r_f_e.support_]
selected_features = list(selected_features) + ['Diabetes_binary']
selectedData = data[selected_features]

print("Selected Features: ", selected_features)
```
